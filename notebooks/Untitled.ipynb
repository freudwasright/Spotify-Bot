```python
import torch
```

```python
# design model -> logistic regression predicts if something is true or false instead of predicting a continous value
# the logistic curves tells the 'probability' of some dependent x and maps it to an independent y(true or false) 
# feedforward loop
# backward pass
# weight updates

class LogisticRegressionBinary(torch.nn.Module):
    def __init__(self, input_dimension, output_dimension):
        super(LogisticRegressionBinary, self).__init__()
        self.Dense = torch.nn.Linear(input_dimension, input_dimension)
        
        
    def forward(self, x):
        x = self.Dense(x)
        x = torch.sigmoid(x)
        return x
```
